```python
def tidy_data(df):
    df_t = df.T.reset_index()
    df_t.columns= df_t.iloc[0]
    df_t= df_t.rename(columns={"Unnamed: 0":"Year"})
    df_t.drop(index= df_t.index[0], axis=0, inplace=True) 
    return df_t
```

```python
def pivot_df(df):
    restruc = pd.melt(df,id_vars= df.columns[0], value_vars= df.columns[1:-1])
    return(restruc)
                     
```
